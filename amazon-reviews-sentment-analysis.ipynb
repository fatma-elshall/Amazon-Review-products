**A general rule of thumb is**:

* A correlation coefficient (r) between 0.1 and 0.3 indicates a weak positive correlation. 

* A correlation coefficient (r) between 0.3 and 0.5 indicates a moderate positive correlation. 

* A correlation coefficient (r) greater than 0.5 indicates a strong positive correlation. 

**Keep in mind that** 

this is just a general guideline and other factors such as sample size, measurement errors, and outliers should also be taken into account when interpreting correlation coefficients.

**Prediction System**

* Based on the previous data and strong correlation between the prices and discounts , we will build a Simple Linear Regression Model independ on the Actual Price and the Discounted Price features .

```python
df.columns
```

```python
# select x and y for the model
x=df['discounted_price'].values.reshape(-1,1)
y= df['actual_price'].values.reshape(-1,1)



#Splitting the dataset into the Training Set and Test Set

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=21)
```

```python
#Fitting Simple Linear Regression to the Training Set

from sklearn.linear_model import LinearRegression

model= LinearRegression()
model.fit(x_train, y_train)
```

```python
#Calculating the Coefficients

model.coef_
```

```python
#Calculating the Intercept

model.intercept_
```

```python
# Make predictions on the testing data
model.predict(x_test)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
#Calculating the R Squared Value
#R2 is a measure that provides information about the goodness of fit of a model .
# it is a statisticl measure of how well the regretion line approximates the actual data.

from sklearn.metrics import r2_score
y_pred = model.predict(x_test)
print('R2 Score: ', r2_score(y_test, y_pred))
```

```python
#Scatter Plot with Regression Line

plt.scatter(x_train, y_train)
plt.plot(x_test, y_pred, color='red')
plt.xlabel('Actual Price', fontweight='bold')
plt.ylabel('Discounted Price', fontweight='bold')
plt.title('Plot with Regression Line', fontweight='bold')
plt.show()
```

* To check for normal distribution of the error, the data points need to be close to the line. From the above plot we can see that there are many data points that are further from the line. This requirement of linear regression is not enough.

* we need more data to build the model becuse these data is very small, **Thus this Simple Linear Regression Model should not be used**.



**sentemnt analysis**

For the Sentiment Analysis, I will be using the VADER Sentiment Scoring. Specifically, using NLTK's Sentiment Intensity Analyzer to get the negative/neutral/positive scores of the text.

```python
#Subsetting Reviews

Reviews= df[['product_id','review_title','review_content']]
```

```python
!pip install twython
```

```python
#importing packages for Sentiment Analysis

import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm

model_2 = SentimentIntensityAnalyzer()
```

```python
#Running the polarity Score on the entire dataset
result = {}

for i, row in tqdm(Reviews.iterrows(), total=len(Reviews)):
    review_desc =row['review_content']
    myid = row['product_id']
    result[myid] = model_2.polarity_scores(review_desc)
```

```python
pol_score = pd.DataFrame(result).T
pol_score = pol_score.reset_index().rename(columns={'index' : 'product_id'})

pol_score
```

```python
#Merging the Dataframe with Polarity Score

Reviwes_df = Reviews.merge(pol_score, how='left')
Reviwes_df.head()
```

```python
#Statistical Description on the Sentiment Score

sentiment_score = Reviwes_df[['pos', 'neu', 'neg', 'compound']]
sentiment_score.describe()[1:7]
```
